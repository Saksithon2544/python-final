```python
import tkinter as tk
from tkinter import messagebox

def calculate():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        operator = operator_var.get()

        if operator == "+":
            result = num1 + num2
        elif operator == "-":
            result = num1 - num2
        elif operator == "*":
            result = num1 * num2
        elif operator == "/":
            if num2 == 0:
                raise ZeroDivisionError
            result = num1 / num2
        else:
            raise ValueError("ไม่พบเครื่องหมายคำนวณ")

        result_label.config(text=f"ผลลัพธ์: {result:.2f}")
    except ZeroDivisionError:
        messagebox.showerror("ข้อผิดพลาด", "ไม่สามารถหารด้วยศูนย์ได้")
    except ValueError:
        messagebox.showerror("ข้อผิดพลาด", "กรุณาป้อนค่าที่ถูกต้อง")

# สร้างหน้าต่างหลัก
root = tk.Tk()
root.title("เครื่องคิดเลข")

# ส่วนรับตัวเลขที่ 1
tk.Label(root, text="ตัวเลขที่ 1:").grid(row=0, column=0, padx=10, pady=10)
entry1 = tk.Entry(root, width=20)
entry1.grid(row=0, column=1, padx=10, pady=10)

# ส่วนเลือกเครื่องหมายคำนวณ
tk.Label(root, text="เครื่องหมาย:").grid(row=1, column=0, padx=10, pady=10)
operator_var = tk.StringVar(value="+")
operator_menu = tk.OptionMenu(root, operator_var, "+", "-", "*", "/")
operator_menu.grid(row=1, column=1, padx=10, pady=10)

# ส่วนรับตัวเลขที่ 2
tk.Label(root, text="ตัวเลขที่ 2:").grid(row=2, column=0, padx=10, pady=10)
entry2 = tk.Entry(root, width=20)
entry2.grid(row=2, column=1, padx=10, pady=10)

# ปุ่มคำนวณ
calculate_button = tk.Button(root, text="คำนวณ", command=calculate)
calculate_button.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

# ส่วนแสดงผล
result_label = tk.Label(root, text="ผลลัพธ์: ", font=("Arial", 12))
result_label.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

# เริ่มต้นโปรแกรม
root.mainloop()
```
